| **TrafficType** | Jenis sumber traffic ke website (1 = langsung, 2 = mesin pencari, dst). |
| **VisitorType** | Tipe pengunjung: Returning_Visitor, New_Visitor, atau Other. Ini fitur penting karena pengunjung lama cenderung lebih besar kemungkinan membeli. |
| **Weekend** | Boolean (True/False) menunjukkan apakah pengunjung datang di akhir pekan. |
| **Revenue** | Target: True jika pengunjung akhirnya melakukan pembelian, False jika tidak. |

## Handling Outlier

```python
df.describe()
```

Menandai outlier dan replace

```python
# List kolom durasi yang mau dicek
duration_cols = ['Administrative_Duration', 'Informational_Duration', 'ProductRelated_Duration']

# Fungsi untuk mendeteksi outlier dengan IQR
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Buat kolom boolean menandai outlier
    outlier_mask = (df[column] < lower_bound) | (df[column] > upper_bound)
    num_outliers = outlier_mask.sum()

    print(f"Outlier di '{column}': {num_outliers} dari total {len(df)} data")
    return outlier_mask

# Cek outlier untuk tiap kolom
outlier_summary = {}
for col in duration_cols:
    outlier_mask = detect_outliers_iqr(df, col)
    outlier_summary[col] = df[outlier_mask]

# Contoh: lihat 5 baris data yang outlier di Administrative_Duration
outlier_summary['Administrative_Duration'].head()
```

```python
def replace_outliers_iqr(df, col_duration, col_count):
    # Hanya ambil data yang count > 0
    valid = df[df[col_count] > 0]

    Q1 = valid[col_duration].quantile(0.25)
    Q3 = valid[col_duration].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    # Hanya ganti nilai outlier di baris dengan count > 0
    outlier_high = (df[col_count] > 0) & (df[col_duration] > upper)
    outlier_low = (df[col_count] > 0) & (df[col_duration] < lower)

    df.loc[outlier_high, col_duration] = upper
    df.loc[outlier_low, col_duration] = lower

    print(f"{col_duration}: Outlier replaced ({outlier_high.sum()} high, {outlier_low.sum()} low)")

# Apply ke 3 kolom
replace_outliers_iqr(df, 'Administrative_Duration', 'Administrative')
replace_outliers_iqr(df, 'Informational_Duration', 'Informational')
replace_outliers_iqr(df, 'ProductRelated_Duration', 'ProductRelated')
```

## Undersampling dan SMOTE

```python
df_encoded = df.copy()
```

```python
from sklearn.model_selection import train_test_split
from collections import Counter

y_target = df_encoded['Revenue'].astype(int)
X_features = df_encoded.drop(columns='Revenue')

# 1. Pisahkan data menjadi set latih dan set uji SEBELUM resampling
X_train, X_test, y_train, y_test = train_test_split(X_features, y_target,
                                                    test_size=0.2,
                                                    random_state=42,
                                                    stratify=y_target)
print(f"Distribusi kelas pada y_train sebelum resampling: {Counter(y_train)}")
print(f"Distribusi kelas pada y_test (tidak akan di-resample): {Counter(y_test)}")
```

```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
from collections import Counter

# SMOTE dulu: buat synthetic minority
over = SMOTE(sampling_strategy=0.5, random_state=42)  # Minority jadi 50% dari majority
# Lalu under sampling: potong majority agar seimbang
under = RandomUnderSampler(sampling_strategy=1.0, random_state=42)  # Seimbangkan 1:1

# Koreksi urutan: over → under
steps = [('o', over), ('u', under)]
pipeline_resampling = Pipeline(steps=steps)

# Terapkan HANYA pada data latih
X_train_resampled, y_train_resampled = pipeline_resampling.fit_resample(X_train, y_train)

# Cek distribusi kelas setelah resampling
print(f"Distribusi kelas pada y_train_resampled: {Counter(y_train_resampled)}")
```

# PEMODELAN

```python
from sklearn.ensemble import RandomForestClassifier

# Filter data train dan test berdasarkan features_selected
X_train_selected = X_train_resampled[top_features]
X_test_selected = X_test[top_features]

# Inisialisasi model
model_rf = RandomForestClassifier(class_weight='balanced' , n_estimators=100, random_state=42)

# Latih model dengan fitur terpilih
model_rf.fit(X_train_selected, y_train_resampled)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns

# Prediksi dengan data uji
y_pred = model_rf.predict(X_test_selected)

# Metrik evaluasi
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")

# Classification report (Precision, Recall, F1-score)
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Visualisasi confusion matrix
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=["False", "True"], yticklabels=["False", "True"])
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
from sklearn.metrics import precision_recall_curve, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import numpy as np

# 1. Dapatkan probabilitas prediksi
y_probs = model_rf.predict_proba(X_test_selected)[:, 1]  # Probabilitas kelas 1

# 2. Cari precision, recall, threshold untuk berbagai cutoff
precision, recall, thresholds = precision_recall_curve(y_test, y_probs)

# 3. Plot Precision-Recall vs Threshold
plt.figure(figsize=(8, 6))
plt.plot(thresholds, precision[:-1], 'b--', label='Precision')
plt.plot(thresholds, recall[:-1], 'g-', label='Recall')
plt.xlabel('Threshold')
plt.ylabel('Score')
plt.title('Precision-Recall vs Threshold')
plt.legend()
plt.grid()
plt.show()
```

```python
threshold = 0.7 # misalnya kamu lihat ini bagus
y_pred_threshold = (y_probs >= threshold).astype(int)

print(classification_report(y_test, y_pred_threshold))
print(confusion_matrix(y_test, y_pred_threshold))
```

```python
from sklearn.model_selection import GridSearchCV # Import GridSearchCV
from sklearn.ensemble import RandomForestClassifier


# Tentukan parameter grid untuk Grid Search
# Perhatikan: untuk GridSearchCV, kita berikan list nilai spesifik, bukan distribusi
param_grid = {
    'n_estimators': [100, 120, 140],       # Variasi di sekitar 103, cenderung sedikit lebih banyak
    'max_depth': [25, 30, 35, None],      # Sekitar 30, dan 'None' untuk melihat tanpa batas
    'min_samples_split': [5, 7, 10],      # Sekitar 7
    'min_samples_leaf': [1, 2, 3],        # Sekitar 2
    'max_features': ['log2', 'sqrt'],     # 'log2' sudah bagus, 'sqrt' sebagai perbandingan umum
    'bootstrap': [False]                  # Fokus pada False karena hasil sebelumnya
}

# Buat model Random Forest
rf = RandomForestClassifier(random_state=42)

# GridSearchCV dengan Cross-Validation
# Perbaikan: Gunakan y_train_resampled alih-alih y_train_selected
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5,
                           scoring='accuracy', n_jobs=-1, verbose=2)

# Lakukan fitting (ini akan memakan waktu lebih lama)
# Perbaikan: Ganti y_train_selected dengan y_train_resampled
grid_search.fit(X_train_selected, y_train_resampled)

# Parameter terbaik yang ditemukan
print("Parameter terbaik:", grid_search.best_params_)

# Akurasi terbaik yang didapatkan
print("Akurasi terbaik:", grid_search.best_score_)

# Model terbaik
best_rf_model = grid_search.best_estimator_
```

```python
import xgboost as xgb
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter # Untuk menghitung scale_pos_weight jika diperlukan

# --- (Opsional tapi Direkomendasikan) Menghitung scale_pos_weight ---
# scale_pos_weight berguna jika data training Anda TIDAK seimbang sempurna,
# atau jika Anda ingin memberi bobot lebih pada kelas positif.
# Jika y_train_resampled Anda sudah 1:1, parameter ini bisa di-set ke 1 atau tidak di-set (default 1).
# Namun, jika Anda ingin menghitungnya berdasarkan distribusi kelas ASLI (sebelum resampling)
# dari data latih awal Anda (misalnya y_train sebelum di-resample):

# Misalkan y_train_original adalah target data latih Anda SEBELUM resampling
# (Jika Anda tidak memiliki variabel ini terpisah, Anda bisa menghitung dari y_target_full sebelum split)
counts_original = Counter(y_train_original) # Ganti y_train_original dengan variabel yang sesuai
scale_pos_weight_value = counts_original[0] / counts_original[1]
# print(f"Calculated scale_pos_weight based on original y_train: {scale_pos_weight_value:.2f}")
# Jika y_train_resampled sudah 1:1, maka scale_pos_weight_value bisa diabaikan atau diset 1.
# Untuk contoh ini, kita akan asumsikan y_train_resampled sudah seimbang dan kita tidak set scale_pos_weight
# atau set ke 1.

# 1. Inisialisasi Model XGBoost Classifier
# Beberapa parameter penting untuk dipertimbangkan (dan di-tuning nanti):
# - n_estimators: Jumlah pohon (boosting rounds)
# - max_depth: Kedalaman maksimum pohon
# - learning_rate (eta): Laju pembelajaran
# - subsample: Fraksi sampel yang digunakan untuk setiap pohon
# - colsample_bytree: Fraksi fitur yang digunakan untuk setiap pohon
# - gamma: Minimum loss reduction untuk membuat split
# - objective: 'binary:logistic' untuk klasifikasi biner (default)
# - eval_metric: Metrik untuk evaluasi selama training (misal 'logloss', 'auc', 'aucpr')
# - scale_pos_weight: Untuk menangani data tidak seimbang (jika y_train_resampled tidak 1:1 atau ingin penekanan)

xgb_model = xgb.XGBClassifier(
    objective='binary:logistic',
    n_estimators=100,       # Bisa ditingkatkan dan di-tune
    learning_rate=0.1,      # Bisa di-tune
    max_depth=5,            # Bisa di-tune
    scale_pos_weight=scale_pos_weight_value, # Gunakan jika y_train_resampled tidak 1:1 atau ingin memberi bobot lebih
    use_label_encoder=False, # Set ke False untuk menghindari warning, karena y sudah numerik
    eval_metric='logloss',   # Atau 'aucpr' jika ingin fokus pada Precision-Recall
    random_state=42
)

# 2. Latih Model
print("Melatih model XGBoost...")
xgb_model.fit(X_train_selected, y_train_resampled,
              # (Opsional) Early stopping:
              # eval_set=[(X_test_selected, y_test)], # Perlu data validasi terpisah, atau X_test
              # early_stopping_rounds=10,
              # verbose=False
             )
print("Model XGBoost selesai dilatih.")

# 3. Lakukan Prediksi pada Data Uji
y_pred_xgb = xgb_model.predict(X_test_selected)
y_pred_proba_xgb = xgb_model.predict_proba(X_test_selected)[:, 1] # Probabilitas untuk kelas positif

# 4. Evaluasi Performa Model
print("\n--- Hasil Evaluasi Model XGBoost pada Data Uji ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred_xgb):.4f}")
# ROC AUC Score penting untuk evaluasi umum model klasifikasi
print(f"ROC AUC Score: {roc_auc_score(y_test, y_pred_proba_xgb):.4f}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred_xgb, target_names=['Revenue False (0)', 'Revenue True (1)']))

print("\nConfusion Matrix:")
conf_matrix_xgb = confusion_matrix(y_test, y_pred_xgb)

plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix_xgb, annot=True, fmt="d", cmap="Blues", xticklabels=["False", "True"], yticklabels=["False", "True"])
plt.title("Confusion Matrix - XGBoost")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
